```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load file
df = pd.read_csv('data.csv')

# 2. Data Cleaning
df['Debt-To-Income Ratio'] = df['Debt-To-Income Ratio'].str.replace('%', '').astype(float)
df['Risk_Score'] = df['Risk_Score'].fillna(df['Risk_Score'].mean())
df['Application Date'] = pd.to_datetime(df['Application Date'], dayfirst=True)

print("Data loaded and cleaned!")

# 3. Top 10 States graph
plt.figure(figsize=(10, 6))
df['State'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 States with Most Loan Requests')
plt.ylabel('Number of Applications')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(df['Risk_Score'], bins=30, kde=True, color='green')
plt.title('Distribution of Risk Scores')
plt.xlabel('Risk Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.sample(1000), x='Risk_Score', y='Amount Requested', alpha=0.5)
plt.title('Relationship between Loan Amount and Risk Score')
plt.xlabel('Risk Score')
plt.ylabel('Amount Requested')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Prepare data
emp_data = df['Employment Length'].value_counts()

# 2. Set figure size and style
plt.figure(figsize=(12, 8))

# 3. Create pie chart
explode = [0.1 if i == '< 1 year' else 0 for i in emp_data.index]
wedges, texts, autotexts = plt.pie(
    emp_data, 
    labels=emp_data.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    explode=explode, 
    pctdistance=0.85, 
    colors=plt.cm.Paired.colors
)

# 4. Adjust font sizes
plt.setp(autotexts, size=10, weight="bold")
plt.setp(texts, size=12)

# 5. Add title
plt.title('Distribution of Employment Experience (Loan Applicants)', fontsize=15, pad=20)

# 6. Add legend
plt.legend(wedges, emp_data.index, title="Experience", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('data.csv')

# Data Cleaning
df['Debt-To-Income Ratio'] = df['Debt-To-Income Ratio'].str.replace('%', '').astype(float)
df['Risk_Score'] = df['Risk_Score'].fillna(df['Risk_Score'].mean())
df['Application Date'] = pd.to_datetime(df['Application Date'], dayfirst=True)

print("Libraries loaded and Data is ready!")

# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.sample(1000), x='Risk_Score', y='Amount Requested', alpha=0.5, color='blue')
plt.title('Relationship between Loan Amount and Risk Score')
plt.xlabel('Risk Score')
plt.ylabel('Amount Requested')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Loan Title'].value_counts().head(10).plot(kind='bar', color='plum')
plt.title('Top 10 Reasons for Loan Applications')
plt.ylabel('Number of People')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Top 10 states with highest average risk score
top_safe_states = df.groupby('State')['Risk_Score'].mean().sort_values(ascending=False).head(10)
print("Top 10 Most Creditworthy States (Average Risk Score):")
print(top_safe_states)

plt.figure(figsize=(10, 5))
top_safe_states.plot(kind='bar', color='green')  # Changed to bar plot for better visualization
plt.title('Top 10 States by Average Risk Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlation_matrix = df[['Amount Requested', 'Risk_Score', 'Debt-To-Income Ratio']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between Financial Metrics')
plt.show()
```

```python
# 1. State mapping dictionary
full_state_map = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 'GA': 'Georgia',
    'HI': 'Hawaii', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa', 'KS': 'Kansas',
    'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland', 'MA': 'Massachusetts',
    'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi', 'MO': 'Missouri', 'MT': 'Montana',
    'NE': 'Nebraska', 'NV': 'Nevada', 'NH': 'New Hampshire', 'NJ': 'New Jersey', 'NM': 'New Mexico',
    'NY': 'New York', 'NC': 'North Carolina', 'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma',
    'OR': 'Oregon', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'SC': 'South Carolina', 'SD': 'South Dakota',
    'TN': 'Tennessee', 'TX': 'Texas', 'UT': 'Utah', 'VT': 'Vermont', 'VA': 'Virginia', 'WA': 'Washington',
    'WV': 'West Virginia', 'WI': 'Wisconsin', 'WY': 'Wyoming', 'DC': 'District of Columbia'
}

# 2. Update State column
df['State'] = df['State'].map(full_state_map).fillna(df['State'])

# 3. Plot top 15 states
plt.figure(figsize=(15, 8))
state_counts = df['State'].value_counts().head(15)
state_counts.plot(kind='bar', color='darkblue', edgecolor='black')
plt.title('Top 15 States by Loan Application Volume', fontsize=16, pad=20)
plt.xlabel('State Full Name', fontsize=12)
plt.ylabel('Number of Applications', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print("States updated!")
```
